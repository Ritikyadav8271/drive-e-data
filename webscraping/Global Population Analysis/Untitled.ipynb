## 🎯 Business Objective
Analyze global population distribution to identify concentration trends.


```python
import requests 
import pandas as pd
from bs4 import BeautifulSoup
from io import StringIO
```

```python
url = "https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_by_population"
```

```python
headers = {
    "User-Agent":"Mozilla/5.0"
}
```

```python
response = requests.get(url,headers = headers)
response.raise_for_status()
```

```python
html_content_io = StringIO(response.text)
```

```python
population_df_list = pd.read_html(html_content_io)
```

```python
population_df = population_df_list[0]
```
