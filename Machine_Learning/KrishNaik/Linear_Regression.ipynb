```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Create sample data
X = np.array([1, 2, 3, 4, 5])      # Example feature (e.g., house size)
Y = np.array([2, 4, 6, 8, 10])     # Actual values (e.g., house price)

# Step 2: Define a function to calculate Mean Squared Error (MSE)
def cost_function(m, c, X, Y):
    predictions = m * X + c
    cost = np.mean((predictions - Y) ** 2)
    return cost

# Step 3: Try different slopes (m) while keeping intercept (c) = 0
m_values = np.linspace(0, 4, 100)   # Slope values from 0 to 4
cost_values = [cost_function(m, 0, X, Y) for m in m_values]

# Step 4: Plot data and example predicted lines
plt.figure(figsize=(14, 6))

# Left plot: data and predicted lines
plt.subplot(1, 2, 1)
plt.scatter(X, Y, color='blue', label='Actual Data')
for m in [0.5, 1.0, 2.0]:
    plt.plot(X, m*X, label=f'Predicted line (m={m})')
plt.title("Linear Regression Predictions with Different Slopes")
plt.xlabel("X (Feature)")
plt.ylabel("Y (Target)")
plt.legend()
plt.grid(True)

# Right plot: cost curve (MSE vs slope)
plt.subplot(1, 2, 2)
plt.plot(m_values, cost_values, color='red')
plt.title("Cost Function Curve (MSE vs Slope m)")
plt.xlabel("Slope (m)")
plt.ylabel("Cost (MSE)")
plt.grid(True)

plt.tight_layout()
plt.show()
```
