```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5,5,400)

f = x**2
g = 3*x**2

plt.figure(figsize=(8,5))
plt.plot(x,f,label='f(x)=x^2')
plt.plot(x,g,label='g(x)=3x^2')
plt.legend()
plt.grid(True)
plt.show()
```

## Scalar  Addictioni

```python
g = x**2 + 3

plt.plot(x,f,label='x^2')
plt.plot(x,g,label='x^2 + 3')
plt.legend()
plt.grid(True)
plt.show()
```

### Sin Function

```python
x = np.linspace(-50,50,500)
y = np.sin(x)

plt.plot(x,y,label='sin(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10,10,200)
y = 2*x + 1

plt.plot(x,y,label='y=2x+1')
plt.grid(True)
plt.legend()
plt.show()
```

```python
x = np.linspace(-10,10,400)

plt.plot(x,np.sin(x),label='sin(x)')
plt.plot(x,np.cos(x),label='cos(x)')
plt.grid(True)
plt.legend()
plt.show()
```
