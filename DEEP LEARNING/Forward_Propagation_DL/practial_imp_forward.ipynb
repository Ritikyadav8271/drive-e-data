```python
import numpy as np
```

```python
# input
X = np.array([[1],[2]])
```

```python
# Layer 1
W1 = np.array([[1,-1],
                [2,1]])
b1 = np.array([[0],[1]])
```

```python
# Layer 2
W2 = np.array([[1,2]])
b2 = -1
```

```python
# forward pass layer 1
z1 = np.dot(W1,X) + b1
a1 = np.maximum(0,z1)  #ReLU
```

```python
z2 = np.dot(W2,a1) + b2
```

```python
# sigmoid apply
y_hat = 1 / (1 + np.exp(-z2))
```

```python
print(y_hat)
```

```python
import matplotlib.pyplot as plt

# Example line: w1*x1 + w2*x2 + b = 0
w1, w2, b = 1, -1, 0

x1 = np.linspace(-10, 10, 100)
x2 = (-w1*x1 - b) / w2


plt.plot(x1, x2)
plt.title("Decision Boundary")
plt.show()
```
