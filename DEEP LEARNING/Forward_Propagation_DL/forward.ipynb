```python
data = {
    "cgpa": [8.5,9.2,7.8],
    "profile_score":[85,92,78],
    "LPA":[10,12,9]
}

df = pd.DataFrame(data)
```

```python
X = df[["cgpa","profile_score"]].values
```

```python
# Initialize Parameter

def initialize_para():
    np.random.seed(1) # fixed integer
    W = np.random.randn(2,1) * 0.01
    b = np.zeros((1,1))
    return W,b
```

```python
def forward_propagation(X,W,b):
    Z = np.dot(X,W) + b
    A = 1 / (1 + np.exp(-Z))
    return A
```

```python
W, b = initialize_para()
A = forward_propagation(X,W,b)
print("Final output: ",A)
```
